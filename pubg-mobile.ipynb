```python
output=data['winPlacePerc']
```

```python
matrix=data.corr()
matrix['winPlacePerc']
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
y=data.pop('winPlacePerc')
x=data
x_train,x_test,y_train,y_test=train_test_split(x,y,shuffle=True)
del x,y,data,matrix
```

```python
model=MLPRegressor(verbose=True,hidden_layer_sizes=(120,120,120,),max_iter=2000)
model.fit(x_train,y_train)
plt.plot(model.loss_curve_)
plt.show()
y_pre1=model.predict(x_test)
y_pre2=model.predict(x_train)
print(mean_squared_error(y_test,y_pre1))
print(mean_squared_error(y_train,y_pre2))
```

```python
import pickle
file=open('model.pickle','wb')
pickle.dump(model,file)
file.close()
```

```python
del x_train,x_test,y_train,y_test,y_pre1,y_pre2
```
